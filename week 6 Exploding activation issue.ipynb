```python
import numpy as np
np.set_printoptions(suppress=True,formatter={'float_kind':'{:0.2f}'.format})
```

```python
x = np.random.rand(1,50)
for i in range(100):
    a = np.random.uniform(-1,1,(50,50))
    x = x@a
    print('std: %.3f  mean: %.3f'% (x.std(),x.mean()))
```

```python
def sigmoid(v):
    return 1/(1+np.exp(-v))

def reLu(v):
    return np.maximum(v,0)

def lReLu(v):
    return np.maximum(v,0.01*v)
```

```python
x = np.random.rand(1,50)
for i in range(100):
    a = np.random.uniform(-1,1,(50,50))*0.3555
    x = reLu(x@a)
    #print(x)
print('%.3f %.3f'%(x.mean(), x.std()))
```
